```python
df=spark.read.csv(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_001.csv", header=True, inferSchema=True)
display(df)
```

##Write DataFrame as Parquet

```python
df.write.save(format="parquet", mode="overwrite", path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output/output_parquet")
```

##Read Parquet Format

```python
df_parquet = spark.read.parquet(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/output/output_parquet"
)
display(df_parquet)
```

##Create a View

```python
df_parquet.createOrReplaceTempView("users_vw")
```

##[Disadvantage] - Updates are not allowed

```python
%sql
UPDATE
    users_vw
SET
    country='Bharat'
WHERE
    country='India'
```

## Disadvantage - Overwrite existing directory no support for ROLLBacks, versioning

```python
from pyspark.sql.functions import col

df.filter(col("country")=="India").write.save(format="parquet", mode="overwrite", path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output/output_parquet")
```

```python
dr_parquet=spark.read.parquet("/Volumes/quickstart_catalog/quickstart_schema/sandbox/output/output_parquet")
dr_parquet.display()
```
