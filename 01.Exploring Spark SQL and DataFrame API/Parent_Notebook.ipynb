#Scenario 01 - Invoke Child notebook

###Invoking child notebook with exit execution message 

```python
execution_status=dbutils.notebook.run("./Child_notebook_01", timeout_seconds=100)
print(execution_status)
```

###Invoking widgets also through arguments from child notebook 

```python
execution_status = dbutils.notebook.run(
    path="./Child_notebook_01",
    timeout_seconds=100,
    arguments={
        "top_n_records": "100",
        "Environment": "Production",
        "name": "Naveen",
        "technology": "4G",
    },
)
print(execution_status)
```
