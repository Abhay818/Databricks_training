#Column Selection & Filtering

##Select only FirstName, LastName, and Salary.


```python
from pyspark.sql.functions import col
df_employee.select(col("FirstName"), col("LastName"), col("Salary")).display()
```

##Show employees with salary greater than 60,000.


```python
df_employee.filter(col("Salary")>60000).select("EmployeeID", "FirstName", "Salary").display()
```

##Find employees who belong to DepartmentID = 1.

```python
df_employee.filter(col("DepartmentID")==1).select("EmployeeID", "FirstName").display()
```

##Filter employees hired after 2021-01-01.

```python
df_employee.filter(col("HireDate")>"2021-01-01").select("EmployeeID", "FirstName", "HireDate").display()
```

##Display employees whose last name starts with "S".

```python
df_employee.filter(col("LastName").startswith("S")).select("EmployeeID", "FirstName", "LastName").display()
```

#Column Operations

##Create a new column AnnualSalary = Salary * 12.


```python
from pyspark.sql.functions import when, col
df_employee.withColumn("Annual Salary", col("Salary")*12).display()
```

##Rename column PhoneNumber to Phone.

```python
from pyspark.sql.functions import col
df_employee.select("*", col("PhoneNumber").alias("Phone")).drop(col("PhoneNumber")).display()
```

##Drop the column Email.

```python
from pyspark.sql.functions import col
df_employee.select("*").drop(col("Email")).display()
```

##Cast Salary column to Integer.

```python
df_employee.select("Salary", col("Salary").cast("int")).display()
df_employee.printSchema()
```

##Add a constant column Country = "India".

```python
from pyspark.sql.functions import lit
df_employee.withColumn("Country", lit("India")).display()
```

#Aggregations

```python
from pyspark.sql.functions import avg
avg_salary=df_employee.select(avg("Salary")).collect()[0][0]
avg_salary
```

```python
from pyspark.sql.functions import avg, max
df_employee.select(max("Salary").alias("max_salary")).display()
```

```python
from pyspark.sql.functions import count
df_employee.groupBy("DepartmentName").agg(count("*").alias("employee_count")).display()
```

##Find total salary per department.

```python
from pyspark.sql.functions import sum
df_employee.groupBy("DepartmentName").agg(
    sum("Salary").alias("Total_Salary")
).display()
```

##Find a department having the highest average salary.

```python
from pyspark.sql.functions import col, avg, max, count, sum, desc
df_employee.groupBy("DepartmentName") \
                                 .agg(avg("Salary").alias("Average_Salary")) \
                                 .orderBy(desc("Average_Salary")) \
                                 .limit(1).display()
```

#Sorting & Distinct
- Sort employees by salary descending.
