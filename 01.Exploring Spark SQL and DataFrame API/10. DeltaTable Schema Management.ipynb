#Prerequisites

```python
%sql
-- CREATE CATALOG IF NOT EXISTS quickstart_catalog;
USE CATALOG quickstart_catalog;
-- CREATE SCHEMA IF NOT EXISTS quickstart_schema;
```

##Load 'user_001.csv' into Delta Table

```python
spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_001.csv",
    header=True,
    inferSchema=True,
).write.mode("overwrite").saveAsTable("quickstart_schema.users_new")
```

#Schema Enforcement


- if we try to add data with different schema then it will throw an error

```python
spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_006_new_column_education.csv",
    header=True,
    inferSchema=True,
).write.mode("append").saveAsTable("quickstart_schema.users_new")
```

#Schema Evolution- Merge schema

```python
spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_006_new_column_education.csv",
    header=True,
    inferSchema=True,
).write.option("mergeSchema", True).mode("append").saveAsTable("quickstart_schema.users_new")
```

```python
from pyspark.sql.functions import col
spark.read.table("quickstart_schema.users_new").filter(col("id").between(1,5)).display()
```
