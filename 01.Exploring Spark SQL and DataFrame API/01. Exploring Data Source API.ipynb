# Custom Schema

```python
from pyspark.sql.types import *

USER_SCHEMA = StructType(
    [
        StructField("id", IntegerType()),
        StructField("age", IntegerType()),
        StructField("gender", StringType()),
        StructField("designation", StringType()),
        StructField("salary", IntegerType()),
    ]
)
# if any datatype is mismatch then it will show as null

df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.csv",
    header=True,
    schema=USER_SCHEMA
    #custom schema is set to optimize the data performance as if there billions of data then custom schema is faster
)
df.display()
```

#DataFrame Writer API


```python
df=spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.csv",
    header=True,
    inferSchema=True,
)
df.write.save(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output/output_json1",
              format="json",
              )

spark.read.json(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output/output_json1"
).display()

#if 
```

#Output Modes

###errorIfExist
- if it will be run again then it will throw error because that path exists

```python
df.write.save(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output/output_json1",
              format="json",
              )
```

###Overwrite
- it will overwrite the existing data

```python
df.write.save(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output/output_json1",
              format="json",
              mode="overwrite"
              )
```

###Append
- It appends the data below in the same file

```python
df.write.save(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output/output_json1",
              format="json",
              mode="append")

spark.read.json(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output/output_json1",
              ).count()
```

###ignore
- It will ignore the data 


```python
df.write.save(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output/output_json1",
              format="json",
              mode="ignore")
```

```python
spark.read.json(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output/output_json1",
              ).count()
```
